### Examining Hayek's writings..

[Lecture 2: The Conditions of Equilibrium between the Production of Consumers’ Goods and the Production of Producers’ Goods](https://cdn.mises.org/prices_and_production_and_other_works.pdf)

Hayek refers to land(e.g. real property) and labor as _**original means of production**_
_**factors of production**_ refers to concepts of wages, rent, interest; to the receipients ("owners") of these factors of production, this is income which can be used to purchase consumer goods.


What are the causes of variation of industrial output? Three primary arguments:
-"...changes in the willingness of individuals to expand effort."

-"...changes of the amount of factors of production used." In this description, _**factors of production**_ refers to concepts of wages, rent, interest. The position suggests that changes in rent, wages and interst are perhaps inversely proportional to  changes in production. Reduction of wages could reflect in more capital available to producers for use in production; contraction in interest rates could also increase capital available to producers. Conversely, increase in wages and interest rates should reduce production capacity. Key point is that idled resources are not required for a company to increase production. I suspect Hayek intends to differentiate unused, or idle, resoruces from inefficient use of resoruces and spare capacity.

-"Changes in the direction given to the existing productive forces are not only the main cause of fluctuations of the output of individual industries; the output of industry as a whole may also be increased or decreased to an enormous extent by changes in the use made of existing resources."

In the third argument, Hayek refers less about _how_ resources are deployed, but more closely how resources are deployed for the future.

## Inflation Break-even Rate

The 5-year inflation break even rate reflects the expected rate of inflation over the next five year; this could be considered the short-term expectation of inflation. Conversely, the 30-year rate is the long-term expected inflation rate and the different between the 5-year and the 30-year provides context for **HOW/WHO IS THIS** view short-term inflation. Positive percentages indicate expected short and near-term inflation.

**DOES THIS MEAN** at the COVID shudown, there was a general expectation of short-term deflation pressures?

```python
start = datetime.datetime(2000,1,1)
end = datetime.datetime.today()
fed_be = ['T5YIE', 'T7YIEM', 'T10YIEM', 'T20YIEM', 'T30YIEM']

# #key/column names
be = {'T5YIE': '5 YR Break-even',
      'T7YIEM': '7 YR Break-even',
      'T10YIEM': '10 YR Break-even', 
      'T20YIEM': '20 YR Break-even', 
      'T30YIEM': '30 YR Break-even'}

breakeven_df = web.DataReader(fed_be, 'fred', start, end)
breakeven_df.columns = breakeven_df.columns.map(be)
breakeven_df.dropna(axis=0, how="any", inplace=True)

```

```python

break_df2 = breakeven_df[['5 YR Break-even', '30 YR Break-even']]
break_df2['Spread'] = break_df2['5 YR Break-even'] - break_df2['30 YR Break-even']
break_df2 = break_df2[['Spread']]

fig, ax = plt.subplots(figsize = (24,16))
break_df2.plot.line(ax=ax, legend=True)
y_vals = ax.get_yticks()
ax.set_yticklabels([str(round(y,2)) + "%" for y in y_vals])
plt.title("% Change in Fed Inflation Breakeven Rate")
```

## Import Dealership Pricing Data
In the code below, I import dealership pricing data from a .csv file. The data was previously downloaded from R. D. Offutt Company's Data Warehouse -- the source data was [Machine Finder Pro](www.machinefinderpro.com). The 'AskingPrice' reflects the dealership's listed retail price on the Machine Finder Pro site. The 'PredictedValue' is result of RDO's Analytics Team's regression algorithm.

The code below:
1. Imports the data
2. Sorts the data by Serial Number ID, then ObservationDate
3. Resets the dataframe index
4. Gets the previously predicted market price (value) for a serial number
5. Calculates the percent of change

```python
#Import
jd_df = pd.read_csv('jd-8series.csv')

#Sort dataframe values
jd_df = jd_df.sort_values(by = (['EquipmentSerialNumberID', 'ObservationDate']))

# Reset the index; (row numbering)
jd_df.reset_index(inplace=True)

# Add column to dataframe using the previous previcted value, but do so within the group
jd_df['prev_predicted_value'] = jd_df.groupby(['EquipmentSerialNumberID'])['PredictedValue'].shift()

# Calculate the % of change
jd_df['pct_change'] = round(((jd_df['PredictedValue'] - jd_df['prev_predicted_value'])/jd_df['prev_predicted_value'] * 100),1)

jd_df
```

## Visualize the Market Direction

Anectdotally, RDO Equipment Sales Management teams contends that used equipment has commanded higher prices than anticipated. One assumes the higher market valuations are a response to limited inventory for new equipment.

the question is can we observe the market direction by plotting the hundreds of unique serial numbers as a line graph where the line is showing the cumulative percentage of gains over the time series. In this visual, the time series is a four month window for which RDO Analtyics team has trained models used for predicting market valuations.

We may also needs to subset the overal plot into various gropuings of equipment. Perhaps 5 year increments, or groupings by horsepower, or within regions.

```python
df_chg = jd_df[['EquipmentSerialNumberID', 'ObservationDate', 'pct_change']]

# Group by serial number & date; calc mean of any in-month chnages
df_chg = df_chg.groupby(['EquipmentSerialNumberID', 'ObservationDate'])['pct_change'].mean().reset_index()

# Pivot data long to wide...
df_chg = df_chg.pivot(index = 'EquipmentSerialNumberID', columns='ObservationDate', values='pct_change')

# Drop April, May, & June as we don't have predictions for those months; first month of change would be July 2021
# drop rows with any NA values; this would indicate it's either added or solde during the July - September
df_chg = df_chg[['2021-07-01', '2021-08-01', '2021-09-01']].dropna()

df_chg
```
