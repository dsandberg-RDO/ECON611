```python
# Defining the date ranges for stock ticker data pulls...
start = datetime.datetime(2018, 1, 1)
end = datetime.datetime.today()

data_dict = {}
stocks = ['DE', 'CNHI', 'AGCO', 'TWI']   
#DE = John Deere | 
# CNHI = CNH Inc. (Case/New Holland) | 
# ACGO = Massey Fergusson; Fendt
# TWI = Titan International; produces wheels, tires, and undercarriage 
# LNN = Lindsay Corp.; irrigation equipment manufacturer; $1.6B market cap
# ALG = Alamo Group; market capitalization $1.8B

#Gather stock price data from Yahoo! Finaance
for stock in stocks:
    data_dict[stock] = web.DataReader(stock, 'yahoo', start, end)

stock_close = pd.concat([df["Close"] for df in data_dict.values()], axis=1, keys = data_dict.keys())

# Producer Price Index 

# Treasury Inflation-Protected Securities:
# https://ycharts.com/indicators/5_year_tipstreasury_breakeven_rate
# 5, 10, 20, 30 Year TIPS data
```

## Import Dealership Pricing Data
In the code below, I import dealership pricing data from a .csv file. The data was previously downloaded from R. D. Offutt Company's Data Warehouse -- the source data was [Machine Finder Pro](www.machinefinderpro.com). The 'AskingPrice' reflects the dealership's listed retail price on the Machine Finder Pro site. The 'PredictedValue' is result of RDO's Analytics Team's regression algorithm.

The code below:
1. Imports the data
2. Sorts the data by Serial Number ID, then ObservationDate
3. Resets the dataframe index
4. Gets the previously predicted market price (value) for a serial number
5. Calculates the percent of change

```python
#Import
jd_df = pd.read_csv('jd-8series.csv')

#Sort dataframe values
jd_df = jd_df.sort_values(by = (['EquipmentSerialNumberID', 'ObservationDate']))

# Reset the index; (row numbering)
jd_df.reset_index(inplace=True)

# Add column to dataframe using the previous previcted value, but do so within the group
jd_df['prev_predicted_value'] = jd_df.groupby(['EquipmentSerialNumberID'])['PredictedValue'].shift()

# Calculate the % of change
jd_df['pct_change'] = round(((jd_df['PredictedValue'] - jd_df['prev_predicted_value'])/jd_df['prev_predicted_value'] * 100),1)

jd_df
```

## Visualize the Market Direction

Anectdotally, RDO Equipment Sales Management teams contends that used equipment has commanded higher prices than anticipated. One assumes the higher market valuations are a response to limited inventory for new equipment.

the question is can we observe the market direction by plotting the hundreds of unique serial numbers as a line graph where the line is showing the cumulative percentage of gains over the time series. In this visual, the time series is a four month window for which RDO Analtyics team has trained models used for predicting market valuations.

We may also needs to subset the overal plot into various gropuings of equipment. Perhaps 5 year increments, or groupings by horsepower, or within regions.

```python
df_chg = jd_df[['EquipmentSerialNumberID', 'ObservationDate', 'pct_change']]

# Group by serial number & date; calc mean of any in-month chnages
df_chg = df_chg.groupby(['EquipmentSerialNumberID', 'ObservationDate'])['pct_change'].mean().reset_index()

# Pivot data long to wide...
df_chg = df_chg.pivot(index = 'EquipmentSerialNumberID', columns='ObservationDate', values='pct_change')

# Drop April, May, & June as we don't have predictions for those months; first month of change would be July 2021
# drop rows with any NA values; this would indicate it's either added or solde during the July - September
df_chg = df_chg[['2021-07-01', '2021-08-01', '2021-09-01']].dropna()

df_chg
```
