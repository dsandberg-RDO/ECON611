### Background on Offutt Farms and LambWestion/RDO

[Offutt Farms](https://www.rdoffuttfarms.com/) is the farming business unit of R. D. Offutt Company, headquartered in Fargo, North Dakota. LambWeston is a agricultural processing company with locations in the Columbia Basin in Nortwest United States, as well as international processing facilities. LambWeston and R. D. Offutt Company joinly own and operate a potato processing facility in Park Rapids, Minnesota.

Offutt Farms plants and harvests roughly 25,000 acres of potatoes, annually, to supply the Park Rapids processing plant. Offutt Farms has dozens of storage warehouses across Minnesota, Wisconsin, and North Dakota where harvested potatoes are stored in climate-controlled facilities unitl they can be processed into frozen potato products. LambWeston holds contracts to supply a variety of retailers with frozen potato products including French fries, hashbrown patties, hashbrowns, puffs(e.g. tater tots), etc. 

For potato processors, a key challenge is selecting the potatoes, meeting specific quality characteristics, from the various storage facilities to meet their retail customers needs. Several of LambWeston's retail customers' qualtity specifications emphasize size of the French fry, blemish free, and a resulting fry color which is close to a pale, golden color. 

### Russet Burbank

Across the potato processing industry, the Russet Burbank cultivar has been a favored variety because of it's physical profile (e.g. density, low sugar content, long, blocky shape) and it's familiarity with producers to grow and store them long-term. There are other Russet cultivars such as Umatilla, Ranger and Alpine which are grown in Minnestoa and the Dakotas; they are grown in smaller numbers (acres) and used to help processors' plans for near-continuous processing.
