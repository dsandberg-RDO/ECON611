```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import pingouin
from datlib.dag import build_skeleton, graph_DAG_skeleton, graph_DAG
```

```python
df = pd.read_csv("../In Class Work/fraserDataWithRGDPPC.csv", index_col = [0,1], parse_dates = True)

df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(axis=0, how="any", inplace=True)

df
```

```python
df.pcorr().round(2)
```

## Homework Question 1
Create a graphical skeleton like the one created above using the Economic Freedom of the World Index and Real GDP.

```python
undirected_graph = {key:[] for key in df.keys()}
for x in undirected_graph:
    remaining_vars = [y for y in df.keys() if y != x]
    for y in remaining_vars:
        undirected_graph[x].append(y)

undirected_graph = build_skeleton(df, undirected_graph)
```

```python
graph_DAG_skeleton(
    undirected_graph,
    df,
    title = "Undirected Graph w/ Partial Correlations\nfrom Full Set of Controls"
)
```

## Homework Question 2
Create a directed acyclic graph using the same data. Include partial correlations as explained in the D-Separation section.

```python
from pgmpy.estimators import PC

c = PC(df)
max_cond_vars = len(df.keys()) - 2

p_val = .01
model = c.estimate(
    return_type = "dag",
    variant= "orig", 
    significance_level = p_val, 
    max_cond_vars = max_cond_vars, 
    ci_test = "pearsonr"
)

edges = model.edges()
```

```python
graph_DAG(edges, df, title = "Directed Acyclic Graph")
```
