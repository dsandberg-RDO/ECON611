```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import pingouin
from datlib.dag import build_skeleton, graph_DAG_skeleton, graph_DAG
```

```python
# Read data into df
df = pd.read_csv("../In Class Work/fraserDataWithRGDPPC.csv", index_col = [0,1], parse_dates = True)

# get list of unique year values; 
# this this materially different using .unique() as .unique() works only on df.values
# and df.index.get_level_values is for indexes; just as says....
# so, #years_unique = np.array(sorted(list(df['Year'].unique()))) doesn't work..
years = np.array(sorted(list(set(df.index.get_level_values("Year")))))

# Now we set a date range; not clear to me at this time why that is necessary
# frequency "AS" = Year Start frequency; 
# frequency parameter values for pandas.date_range: 
# https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#timeseries-offset-aliases
# looks like we're using years[-2] becasue the last year will be the end of the date range between years[-2] and years[-1]
years = pd.date_range(years[0], years[-2], freq="AS") 

#List of unique levels of ISO_Codes from index
countries = np.array(sorted(list(set(df.index.get_level_values("ISO_Code")))))

# Looks like this will be used to rename the indexes in teh reindex statement
index_names = list(df.index.names)

# Looks like a cross-join (Cartesian join to produce all combinations...)
multi_index = pd.MultiIndex.from_product([countries,years[:-1]], names = index_names)

df = df.reindex(multi_index)
```

```python
print(list(df.keys()))
```

```python
#Not certain why we're omitting the first key [0]...
keys = list(df.keys())[1:]

dx = df.copy()
dx['RGDP Per Capita'] = df['RGDP Per Capita'].groupby('ISO_Code').pct_change()

# This is the difference of all keys?
dx = dx[keys].diff()

dx = dx.rename(columns={key:key.replace(" ", "\n") for key in dx})

# Now, drop NAs
dx.dropna(how="any", axis = 0, inplace=True)
```

```python
# Get partial correlations... 
dx.pcorr().round(2)
```

## Homework Question 1
Create a graphical skeleton like the one created above using the Economic Freedom of the World Index and Real GDP.

```python
undirected_graph = {key:[] for key in dx.keys()}
for x in undirected_graph:
    remaining_vars = [y for y in dx.keys() if y != x]
    for y in remaining_vars:
        undirected_graph[x].append(y)

undirected_graph = build_skeleton(dx, undirected_graph)
```

```python
graph_DAG_skeleton(
    undirected_graph,
    dx,
    title = "Undirected Graph w/ Partial Correlations\nfrom Full Set of Controls"
)
```

## Homework Question 2
Create a directed acyclic graph using the same data. Include partial correlations as explained in the D-Separation section.

```python
from pgmpy.estimators import PC

c = PC(dx)
max_cond_vars = len(dx.keys()) - 2

p_val = .01
model = c.estimate(
    return_type = "dag",
    variant= "orig", 
    significance_level = p_val, 
    max_cond_vars = max_cond_vars, 
    ci_test = "pearsonr"
)

edges = model.edges()
```

```python
graph_DAG(edges, dx, title = "Directed Acyclic Graph")
```
