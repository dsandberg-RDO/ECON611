### Authoring Function to Create "Indicator" Variables

Don't know what Prof. Caton means by indicator variables... need to better understand, or translate what an indicator vaialble is....

"indicator" is simply setting a new column as a bit/binary, then applying to rule to another set oc daeta to determine if the value in that row/column **IS** OR **IS NOT** meeting the rule.  So, Is this crop a Burbank variety crop or not.. then add a new column called IsBurbank, as set the values as 0 / 1.

```python
def create_indicator_variable(data, indicator_name, index_name, target_index_list):  
    #set default value to 0; in other cases you might chose True/1... depending on what you're doing
    data[indicator_name] = 0  
    # set the desired items to 1/True!
    data.loc[target_index_list, [indicator_name]] = int(1)
```

```python
# select "ISO_Code" from names of double index  
index_name = data.index.names[0]  
indicator_name = "North America"  
# Cuba, Grenada, Saint Kitts, Saint Lucia, Saint Vincent are missing   
# from Fraser Data  
countries_in_north_america = [
    "BHS", "BRB", "BLZ", "CAN", "CRI", "DOM", "SLV",
    "GTM", "HTI", "HND", "JAM", "MEX", "NIC", "PAN",
    "TTO", "USA"]  
create_indicator_variable(data = data, 
    indicator_name = indicator_name,  
    index_name = index_name, 
    target_index_list = countries_in_north_america)
```
