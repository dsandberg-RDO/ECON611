```python
two_dim_list = [[2,3,4,5], [9,8,7,6], [1,3,6,9]]

for outer in two_dim_list:
    print(outer)
    for inner in outer:
        print(inner)
```

## Monte Carlo Sim

```python
# build a list of random values sampled from a normal distribution with a mean of 1.2 and sigma = .088
random_list = []
list_len = 1000
mean = 0
sigma = .5
for i in range(list_len):
    random_list.append(random.normalvariate(mean, sigma))
plt.plot(random_list, ls = "", marker = ".", markersize = 1.5)
plt.show()
plt.close()
```

```python
simulation = []
for i in range(len(random_list)):
    val = random_list[i]
    if i == 0:
        simulation.append(val)
    else:
        simulation.append(simulation[i-1] + val)
        
plt.plot(simulation)
plt.show()
plt.close()
```

```python
mcdict = {}
sum_sims = 2000
periods = 1000
for i in range(sum_sims):
    mcdict[i] = {}
    for period in range(periods):
        curr_sim = mcdict[i]
        if period == 0:
            curr_sim[period] = random.normalvariate(mean, sigma)
        else:
            curr_sim[period] = curr_sim[period-1] + random.normalvariate(mean, sigma)
            
mcdict
mc_df = pd.DataFrame(mcdict)
print(mc_df)
```

```python
fig, ax = plt.subplots(figsize = (14,8))
mc_df.plot.line(ax=ax,
               color = "k",
               alpha = .01,
               legend = False)
plt.show()
plt.close()
```

### Functions

```python
def total(list):
    total_ = 0
    for val in list:
        total_ += val
    return total_

def mean(list):
    n = len(list)
    mean_ = total(list) / n
    return mean_
```

```python
x = [1,3,5,9,11,13,3,5]
```

```python
print("total:",total(x))
print("Mean:", mean(x))
```
